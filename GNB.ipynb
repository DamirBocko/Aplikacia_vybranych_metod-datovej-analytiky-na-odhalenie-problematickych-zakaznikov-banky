```python
from mylibraries import *
X_train = pd.read_csv('X_train_scaled.csv')
X_test = pd.read_csv('X_test_scaled.csv')
y_train = pd.read_csv('y_train.csv')
y_test = pd.read_csv('y_test.csv')

y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

# Získanie názvov stĺpcov z pôvodného X_train
column_names = X_train.columns
```

```python
from mylibraries import *
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Inicializácia a trénovanie modelu Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predikcia na testovacej množine
y_pred = gnb.predict(X_test)

# Vyhodnotenie modelu
accuracy = accuracy_score(y_test, y_pred)
print(f"Presnosť modelu: {accuracy:.4f}")

# Klasifikačná správa
print("Klasifikačná správa:")
print(classification_report(y_test, y_pred))

# Matica zámen
print("Matica zámen:")
print(confusion_matrix(y_test, y_pred))


from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score, precision_recall_curve, average_precision_score
from sklearn.model_selection import learning_curve

# ROC krivka
y_prob_gnb = gnb.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob_gnb)
plt.plot(fpr, tpr, label='ROC krivka (AUC = {:.2f})'.format(roc_auc_score(y_test, y_prob_gnb)))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('Falošne pozitívna miera')
plt.ylabel('Pravdivo pozitívna miera')
plt.title('ROC Krivka pre Gaussian Naive Bayes')
plt.legend()
plt.show()

# Precision-Recall krivka
precision, recall, _ = precision_recall_curve(y_test, y_prob_gnb)
average_precision = average_precision_score(y_test, y_prob_gnb)
plt.plot(recall, precision, label='Precision-Recall krivka (priemer = {:.2f})'.format(average_precision))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Krivka pre Gaussian Naive Bayes')
plt.legend()
plt.show()

# Learning Curve
train_sizes, train_scores, test_scores = learning_curve(gnb, X_train, y_train, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))
train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.figure()
plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Trénovacia presnosť')
plt.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Validačná presnosť')

plt.xlabel('Počet trénovacích príkladov')
plt.ylabel('Presnosť')
plt.title('Learning Curve pre Gaussian Naive Bayes')
plt.legend(loc="best")
plt.show()

# Dôležitosť atribútov
feature_importances = gnb.theta_[0]  # Pre Gaussian Naive Bayes, dôležitosť atribútov je uložená ako priemer hodnôt atribútov v triede 0
sorted_idx = np.argsort(feature_importances)[::-1]
top_n = 10
sorted_idx_top_n = sorted_idx[:top_n]

plt.barh(range(top_n), feature_importances[sorted_idx_top_n])
plt.yticks(range(top_n), column_names[sorted_idx_top_n])
plt.xlabel('Dôležitosť atribútov')
plt.title('Top {} najdôležitejších atribútov pre Gaussian Naive Bayes'.format(top_n))
plt.show()

```

```python
# Vyber iba najdôležitejšie atribúty
top_features = column_names[sorted_idx_top_n]

# Použi iba vybrané atribúty na trénovanie a testovanie
X_train_top = X_train[top_features]
X_test_top = X_test[top_features]

# Inicializácia a trénovanie modelu Gaussian Naive Bayes s najdôležitejšími atribútmi
gnb_top = GaussianNB()
gnb_top.fit(X_train_top, y_train)

# Predikcia na testovacej množine s najdôležitejšími atribútmi
y_pred_top = gnb_top.predict(X_test_top)

# Vyhodnotenie modelu s najdôležitejšími atribútmi
accuracy_top = accuracy_score(y_test, y_pred_top)
print(f"Presnosť modelu s najdôležitejšími atribútmi: {accuracy_top:.4f}")

# Klasifikačná správa s najdôležitejšími atribútmi
print("Klasifikačná správa s najdôležitejšími atribútmi:")
print(classification_report(y_test, y_pred_top))

# Matica zámen s najdôležitejšími atribútmi
print("Matica zámen s najdôležitejšími atribútmi:")
print(confusion_matrix(y_test, y_pred_top))
```

```python
from sklearn.ensemble import BaggingClassifier

# Inicializácia modelu Gaussian Naive Bayes
gnb = GaussianNB()

# Inicializácia modelu Bagging Classifier s Gaussian Naive Bayes ako základným odhadcom
bagging_gnb = BaggingClassifier(estimator=gnb, n_estimators=10, random_state=42)

# Trénovanie modelu
bagging_gnb.fit(X_train, y_train)

# Predikcia na testovacej množine
y_pred_bagging = bagging_gnb.predict(X_test)

# Vyhodnotenie modelu
accuracy_bagging = accuracy_score(y_test, y_pred_bagging)
print(f"Presnosť modelu Bagging Classifier s Gaussian Naive Bayes: {accuracy_bagging:.4f}")

# Klasifikačná správa
print("Klasifikačná správa modelu Bagging Classifier s Gaussian Naive Bayes:")
print(classification_report(y_test, y_pred_bagging))

# Matica zámen
print("Matica zámen modelu Bagging Classifier s Gaussian Naive Bayes:")
print(confusion_matrix(y_test, y_pred_bagging))
```
